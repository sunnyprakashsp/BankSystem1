```python
from sentence_transformers import SentenceTransformer
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
data = [
    {"name": "Boho Dress", "desc": "Flowy, earthy tones for festival vibes"},
    {"name": "Urban Jacket", "desc": "Sleek leather look for city nights"},
    {"name": "Cozy Sweater", "desc": "Soft knit in warm hues for relaxed evenings"},
    {"name": "Athleisure Set", "desc": "Energetic, sporty look for on-the-go comfort"},
    {"name": "Minimalist Blazer", "desc": "Clean lines and neutral tones for modern professionals"},
]
```

```python
df = pd.DataFrame(data)
```

```python
df["embedding"] = df["desc"].apply(lambda x: model.encode(x).tolist())
```

```python
def vibe_matcher(query, df, top_n=3):
    query_emb = model.encode(query)
    embeddings = np.vstack(df["embedding"].values)
    sims = cosine_similarity([query_emb], embeddings)[0]
    df["similarity"] = sims
    top = df.sort_values("similarity", ascending=False).head(top_n)
    return top[["name", "desc", "similarity"]]
```

```python
query = "energetic urban chic"
top_matches = vibe_matcher(query, df)
print(top_matches)
```
